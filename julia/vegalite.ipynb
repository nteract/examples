# Displaying vega-lite with Julia

We use the [VegaLite.jl](https://github.com/queryverse/VegaLite.jl) package to create the figure, and the [VegaDatasets.jl](https://github.com/queryverse/VegaDatasets.jl) package to load the example data used for this plot.

```julia
using VegaLite, VegaDatasets

dataset("cars") |>
@vlplot(
    :point,
    x=:Horsepower,
    y=:Miles_per_Gallon,
    color=:Origin,
    width=400,
    height=400)
```
